```python
import pandas as pd
```

```python
orders = pd.read_csv("../data/Orders-Table 1.csv")
```

```python
orders.groupby("Category")["Sales"].sum()
```

```python
orders.groupby("Category")[["Sales", "Profit"]].sum()
```

```python
orders["Profit Margin"] = orders["Profit"] / orders["Sales"]
```

```python
orders.groupby("Category")["Profit Margin"].mean()
```

```python
orders.groupby("Customer ID")["Sales"].sum()
```

```python
orders.groupby("Customer ID")["Sales"].sum().sort_values(ascending=False)
```

```python
orders.groupby("Customer ID")["Profit"].sum().sort_values(ascending=False)
```

```python
top_customers = orders.groupby("Customer ID")["Profit"].sum().sort_values(ascending=False).head(5)
top_customers
```

```python
top_customer_orders = orders[orders["Customer ID"].isin(top_customers.index)]
top_customer_orders.head()
```

```python
top_customer_orders.groupby("Category")["Sales"].sum()
```

```python
orders["Order Date"] = pd.to_datetime(orders["Order Date"])
```

```python
orders["YearMonth"] = orders["Order Date"].dt.to_period("M")
```

```python
orders.groupby("YearMonth")[["Sales", "Profit"]].sum()
```

```python
monthly_data = orders.groupby("YearMonth")[["Sales", "Profit"]].sum()

monthly_data["Profit Margin"] = monthly_data["Profit"] / monthly_data["Sales"]

monthly_data
```

```python
orders.groupby("Discount")[["Sales", "Profit"]].mean()
```

```python
discount_analysis = orders.groupby("Discount")[["Sales", "Profit"]].mean()

discount_analysis["Profit Margin"] = discount_analysis["Profit"] / discount_analysis["Sales"]

discount_analysis
```

```python
category_discount_analysis = (
    orders.groupby(["Category", "Discount"])[["Sales", "Profit"]]
    .mean()
    .reset_index()
)

category_discount_analysis["Profit Margin"] = (
    category_discount_analysis["Profit"] / category_discount_analysis["Sales"]
)

category_discount_analysis
```

```python
orders.groupby(["Category", "Discount"])["Order ID"].count().reset_index()
```

```python
pd.pivot_table(
    orders,
    values="Order ID",
    index="Discount",
    columns="Category",
    aggfunc="count",
    fill_value=0
)
```

```python
# Step 1: remove 0% discount rows
filtered_orders = orders[orders["Discount"] > 0]

# Step 2: create pivot table
pivot = pd.pivot_table(
    filtered_orders,
    values="Order ID",
    index="Discount",
    columns="Category",
    aggfunc="count",
    fill_value=0
)

# Step 3: add totals column
pivot["Total"] = pivot.sum(axis=1)

# Step 4: add totals row
pivot.loc["Total"] = pivot.sum()

pivot
```
